##Introduction

   Government, business, and non-profit organizations in the United States historically have worked together to achieve important public purposes.  Today, such cross-sector collaborations are more important than ever in addressing the increasing number of complex public issues that spill over sectoral boundaries.  Collaboration can be defined as the sharing of information, resources, activities, and capabilities by organizations to achieve jointly an outcome that could not be achieved by the organizations separately. It occurs because we live in a shared-power world in which many groups and organizations are involved in, affected by, or have some partial responsibility to act on public challenges (Crosby & Bryson, 2005). 
    
   Organizations in each sector have different capabilities and resources to contribute to the problem through cross-sector collaboration. However, the power sharing in a collaboration does not imply equal power, nor does it necessarily imply shared interests and goals. Indeed, cross-sector collaboration typically involves uneven power and mixed motives. 

   Furthermore, government, business and civil society actors differ in how they organise to pursue common goals, mobilise and deploy resources, make decisions, act and are held to account for their actions. Blending these distinct organisational systems to underpin effective collaboration presents considerable challenges - failing to do being the main explanation for shortfalls in partnership outcomes, and vice versa. Diversity in the interaction of these actors across countries, regions and cultures deepens complexities and the challenge of success, especially for international collaborative initiatives. Evolution of the norms informing these actors, furthermore, present exciting opportunities, especially where innovation drives the erosion of differences. By disrupting norms, however, such evolution can also dilute coherence, create confusion, and reduce effectiveness unless understood and carefully handled.

   How to communicate these differences, balance diverse goals and demands, and motivate all parties to respond collaboratively and effectively to problems is a major challenge. The purpose of this study is to explore effective ways that government, business, and nonprofit organizations can align interests and form collaborate through computational modeling. To explore this question, I want to focus specifically on the effect of facilitator in forming collaboration. Instead of just having organizations randomly paired up and check for their requirements for collaboration individually, my hypothesis is that having a facilitator who can pair up organizations that have relatively high degree of commitment will increase the total number of collaboration formed.

# Methods

I. Actors:
A. Organizations
Properties
1. Organization types: government agency, company,and nonprofit organization. 
2. Organization id: each organization/ node has a unique id.
3. Resources A,B,C:
--Each organization has Resources A, B, C. Based on each organizaton's type, its amount of resources A,B,C are different. For government organizations, they will draw values from specific ranges for all resources A,B,C. This apply to business organizatoins and nonprofit organizations. I set different ranges for different types of organizations because I want the model to reflect the reality that organizations in different sector has different amount of resources. 
4. Degree of Commitment (DOC):
--Each organization has degree of commitment, which means an organization's level of commitment to collaborate and solve the problem. Its value is drawn from a range between 0 and 1. 

B. The effect of facilitator
The facilitator in this model is not a actual node, but it play a role in pairing up organizations that have degree of commitment that is equal to and larger than 0.5. 

II. Initial Conditions--Without the effect of facilitator
A. Organizations
Organizations will be created in the network. The number of government agencies, businesses, and nonprofit organizations will be assigned when I call the model. 
Organization will have their resources A,B,C randonly initialized to a value from different ranges which are determined by their organizational types. 
Organizations will have their degree of commitment randomly initialized to a value from a uniform continuous distribution.

B. Facilitator
A facilitator will create an effect on how organizatoins are selected and paired up. It select organizations of different type and they both have degrees of commitement greater than 50%.

III. Model Parameters

●  num_gov: number of government agencies

●   num_bus: number of companies

●   num_np: number of nonprofit organizations

●   gov_AMin, gov_AMax: the lower and upper bounds for the resource A of government agency

●   gov_BMin, gov_BMax: the lower and upper bounds for the resource B of government agency

●   gov_CMin, gov_CMax: the lower and upper bounds for the resource C of government agency

●   gov_DOCMin, gov_DOCMax: the lower and upper bounds for the degree of commitment of government agency

●   bus_AMin, bus_AMax: the lower and upper bounds for the resource A of company

●   bus_BMin, bus_BMax: the lower and upper bounds for the resource B of company

●   bus_CMin, bus_CMax: the lower and upper bounds for the resource C of company

●   bus_DOCMin, bus_DOCMax: the lower and upper bounds for the degree of commitment of company

●   np_AMin, np_AMax: the lower and upper bounds for the resource A of nonprofit organization

●   np_BMin, np_BMax: the lower and upper bounds for the resource B of nonprofit organization

●   np_CMin, np_CMax: the lower and upper bounds for the resource C of nonprofit organization

●   np_DOCMin, np_DOCMax: the lower and upper bounds for the degree of commitment of nonprofit organization
    
IV. Step Methods:

A. Condition without the effect of facilitator.

1. Randomly pair up two organizations, org1 and org2. 

2. Make org1 and org2 to do the condition checking.
   1) if org1 and org2 are the same type, respond return and select organizations again.
   2) if org1 have degree of commitment smaller than 0.5, respond return and select organizations again.
   3) if org2 have degree of commitment smaller than 0.5, respond return and select organizations again.
   
3. Otherwise, continue the condition checking.
   if org1's resource A + org2's resource A >= 1.0
   or if org1's resource B + org2's resource B >= 1.0
   or if org1's resource C + org2's resource C >= 1.0
   
4. If one of the conditions above hold true, than decide collaborate. 
   
5. Run the model for 101 times and measure the number of collaboration formed. 

B. Condition with the effect of a facilitator.

1. With facilitator, instead of randomly selecting two organizations, we can only select pair of organizations that are "good pair", which means:
   1) They are of different type AND  
   2) Have degrees of commitement greater than 50%

2. After the pair are selected, then conduct the condition checking. 
   if org1's resource A + org2's resource A >= 1.0
   or if org1's resource B + org2's resource B >= 1.0
   or if org1's resource C + org2's resource C >= 1.0
   
3. If one of the conditions above hold true, than decide collaborate. 
   
4. Run the model for 101 times and measure the number of collaboration formed. 

V. Meausure and Analyze Result
1. Measure the number of collaboration formed in condition that has no effect of facilitator and condition with the effect of facilitator. Also, I plan to study the organizations that will successfully form collaboration by anaylzing their organization type, resources level, and commitment level. 

**Sweep parameters:

Add the effect of facilitator. Measure the outcome: number of collaboration formed. Compare this number with the number of collaboration formed in a condition that has no facilitator and condition with one facilitator. 
 


```python
%matplotlib inline

# Imports
import copy
import itertools
import networkx as nx
import numpy
import matplotlib.pyplot as plt
import pandas

# Import widget methods
from IPython.html.widgets import *
```

##  Organization Class
●   Constructor:  class constructor, which "initializes" or "creates" the organization when we call org(). The org has five main variables, which are organizational type, organization's id, resource A, resource B, resource C, and degree of commitment. 
●   def print_org: 

```python
class org(object):

    """
    Organizations class, which encapsulates the entire behavior of gov, bus, np 
    """
    
    def __init__(self, model, org_type, org_id, A, B, C, DOC):
        
        # Set model link and ID
        self.model = model
        self.org_type = org_type 
        self.org_id = org_id
        
        # Set Org parameters
        self.A = A
        self.B = B
        self.C = C
        self.DOC = DOC
        
    def print_org(self):
        print(self.org_type, 'A=', round(self.A,2)*100, "B= ", round(self.B,2)*100, "C= ", 
              round(self.B,2)*100, "DOC= ", round(self.DOC,2)*100) 
```

## Model Class

1.constructor: class constructor, which "initializes" or "creates" the model when we call Model(). This is in the __init__ method.
2.setup_org: method to create organizations.
3.step_select: method to select org1 and org2 randomly without the effect of facilitator.
4.step_select_with_facilitator: select org1 and org2 randomly with the effect of facilitator.
5.step: main step method to control each time step simulation without the effect of facilitator.
6.step_with_facilitator: main step method to control each time step simulation with the effect of facilitator. 



```python
class Model(object):
    """
    Model class, which encapsulates the entire behavior of a single "run" in this model.
    """
    
    def __init__(self, num_gov=3, num_bus=3, num_np=3, gov_AMin=0.4, gov_AMax=0.6, gov_BMin=0.0, gov_BMax=0.2, 
                 gov_CMin=0.4,gov_CMax=0.6, gov_DOCMin=0.0, gov_DOCMax=1.0, bus_AMin=0.4, bus_AMax=0.6, 
                 bus_BMin=0.8, bus_BMax=1.0, bus_CMin=0.0, bus_CMax=0.2, bus_DOCMin=0.0, bus_DOCMax=1.0, 
                 np_AMin=0.0, np_AMax=0.2, np_BMin=0.4, np_BMax=0.6, np_CMin=0.8, np_CMax=1.0,
                 np_DOCMin=0.0, np_DOCMax=1.0):
        """
        Class constructor.
        """
        # Set our model parameters; this is long, and also hard!
        self.num_gov = num_gov
        self.num_bus = num_bus
        self.num_np = num_np
        self.gov_AMin = gov_AMin
        self.gov_AMax = gov_AMax
        self.gov_BMin = gov_BMin
        self.gov_BMax = gov_BMax
        self.gov_CMin = gov_CMin
        self.gov_CMax = gov_CMax
        self.gov_DOCMin = gov_DOCMin
        self.gov_DOCMax = gov_DOCMax
        self.bus_AMin = bus_AMin
        self.bus_AMax = bus_AMax
        self.bus_BMin = bus_BMin
        self.bus_BMax = bus_BMax
        self.bus_CMin = bus_CMin
        self.bus_CMax = bus_CMax
        self.bus_DOCMin = bus_DOCMin
        self.bus_DOCMax = bus_DOCMax
        self.np_AMin = np_AMin
        self.np_AMax = np_AMax
        self.np_BMin = np_BMin
        self.np_BMax = np_BMax
        self.np_CMin = np_CMin
        self.np_CMax = np_CMax
        self.np_DOCMin = np_DOCMin
        self.np_DOCMax = np_DOCMax
        
         # Set our state variables
        self.t = 0
        self.orgs = []
        self.num_interactions = 0
        self.num_collaborations = 0
        self.G5 = nx.Graph()
        
        # Setup our history variables.
        self.history_interactions = []
        self.history_collaborations = []
        self.history_num_collaborations = []
        self.history_collaborating_orgs = []
        self.history_freq_collaborating_orgs = []
            
        # Call our setup methods to initialize space, organizations
        self.setup_org()
        
      
    def setup_org(self):
        """
        Create organizations
        """
        # create government agencies.
        for i in xrange(self.num_gov):
            self.orgs.append(org(model=self,
                                org_id=i,
                                org_type = "gov",
                                A=numpy.random.uniform(self.gov_AMin, self.gov_AMax),
                                B=numpy.random.uniform(self.gov_BMin, self.gov_BMax),
                                C=numpy.random.uniform(self.gov_CMin, self.gov_CMax),
                                DOC=numpy.random.uniform(self.gov_DOCMin, self.gov_DOCMax)))
       
            
        for i in xrange(self.num_bus):
            self.orgs.append(org(model=self,
                                org_id=i+self.num_gov,
                                org_type = "bus",
                                A=numpy.random.uniform(self.bus_AMin, self.bus_AMax),
                                B=numpy.random.uniform(self.bus_BMin, self.bus_BMax),
                                C=numpy.random.uniform(self.bus_CMin, self.bus_CMax),
                                DOC=numpy.random.uniform(self.bus_DOCMin, self.bus_DOCMax)))

            
        for i in xrange(self.num_np):
            self.orgs.append(org(model=self,
                                org_id=i+self.num_bus+self.num_gov,
                                org_type = "np",
                                A=numpy.random.uniform(self.np_AMin, self.np_AMax),
                                B=numpy.random.uniform(self.np_BMin, self.np_BMax),
                                C=numpy.random.uniform(self.np_CMin, self.np_CMax),
                                DOC=numpy.random.uniform(self.np_DOCMin, self.np_DOCMax)))
       

    def print_collaborating_org(self):
        """
        Create the value of the table of the organizations that collaborate. I will print the table after I test the model.
        """
        print_vals = []
        ii = 0
        for index in self.history_collaborating_orgs:
            ooorg = self.orgs[index]
            print_vals.append((self.history_freq_collaborating_orgs[ii], ooorg.org_id,
                  ooorg.org_type, round(ooorg.A*100), round(ooorg.B*100), 
                  round(ooorg.C*100), round(ooorg.DOC*100)))
            ii += 1
        return print_vals
    
    
    def step_select(self):       
        """
        Select org1 and org2 randomly without the effect of facilitator. 
        """
        node_id = xrange(self.num_np + self.num_gov + self.num_bus)
        [node1, node2] = numpy.random.choice(node_id, size=2, replace=False)
        
        org1 = self.orgs[node1]
        org2 = self.orgs[node2]
        
    # Store the pair that was selected and its symmetric so that they are not selected again.
    
        if (node1, node2) not in self.history_interactions:
             self.history_interactions.append((node1, node2)) 
             self.history_interactions.append((node2, node1))
        else:
             return
           
        if org1.org_type == org2.org_type or org1.DOC < 0.5 or org2.DOC < 0.5:
            return
        else:            
            if org1.A + org2.A >= 1.0 or org1.B + org2.B >=1.0 or org1.C + org2.C >=1.0 :  
                self.history_collaborations.append((node1, node2))  
        
         # Store the two organizations that fulfilled the collaboration conditions; 
        # In addition store the number of times each of the organizations collaborated
        
                for node in [node1,node2]:
                    if node not in self.history_collaborating_orgs:
                        self.history_collaborating_orgs.append(node)
                        self.history_freq_collaborating_orgs.append(1)
                    else:
                        index_node = self.history_collaborating_orgs.index(node)
                        self.history_freq_collaborating_orgs[index_node]+=1
                        
                        
    def step_select_with_facilitator(self):  
        """
        Select org1 and org2 randomly with the effect of facilitator. 
        """
        node_id = xrange(self.num_np + self.num_gov + self.num_bus)
        
        # with facilitator, we keep selecting pairs of nodes until we find a good pair within the time step
        # A 'good' pair is one for which the organizations
        # 1) Are of different type AND  
        # 2) Have degrees of commitement greater than 50%
        
        good_pair = False
        while(not good_pair):
            [node1, node2] = numpy.random.choice(node_id, size=2, replace=False)
        
            org1 = self.orgs[node1]
            org2 = self.orgs[node2]
            
            if not(org1.org_type == org2.org_type or org1.DOC < 0.5 or org2.DOC < 0.5):
                good_pair = True
        
        if (node1, node2) not in self.history_interactions:
             self.history_interactions.append((node1, node2)) 
             self.history_interactions.append((node2, node1))
        else:
             return
           
        
        if org1.A + org2.A >= 1.0 or org1.B + org2.B >=1.0 or org1.C + org2.C >=1.0:  
            self.history_collaborations.append((node1, node2))  
                
            for node in [node1,node2]:
                if node not in self.history_collaborating_orgs:
                    self.history_collaborating_orgs.append(node)
                    self.history_freq_collaborating_orgs.append(1)
                else:
                    index_node = self.history_collaborating_orgs.index(node)
                    self.history_freq_collaborating_orgs[index_node]+=1
                                           
                    
    def step(self):
        """
        Model step function.
        """
        self.step_select()

        self.t += 1
        self.num_interactions = len(self.history_interactions)/2
        self.num_collaborations = len(self.history_collaborations)
        self.history_num_collaborations.append(self.num_collaborations)
        
    def step_with_facilitator(self):
        """
        Model step function (With the effect of facilitator)
        """
        self.step_select_with_facilitator()

        self.t += 1
        self.num_interactions = len(self.history_interactions)/2
        self.num_collaborations = len(self.history_collaborations)
        self.history_num_collaborations.append(self.num_collaborations)
         
    
```

## Testing our model

In order to test our model, we can:
1.Initialize two models, both with and without the effect of facilitator.
2. Assign value to some parameters, such as num_gov,num_bus, num_np.
3. Iterate for some numner of steps. For both models, I take 101 steps, outputting some information every 100 steps.
4. Plot some important statistics as tables and time series plots. 


```python
# Create model and output
m = Model(20,20,20)
# Step over the model for a few steps
for i in xrange(101):
    m.step()  
    # Update every 100 steps
    if i % 100 == 0:
        print(m.t, m.num_interactions, m.num_collaborations, m.history_collaborations)
```

Secondly, with the effect of facilitator, the number of collaboration formed increase significantly. According to the first time series graph, there are only 14 collaboration formed over 101 time steps without the effect of facilitator. However, if we add the effect of facilitator, in another word, selecting pair of organizations that are of different type and have degrees of commitement greater than 50%; we get 80 collaboration formed over 101 time steps. This outcome is 82.5% higher than the number of collaboration formed without the effect of facilitator. 

These findings suggests that if an organization wants to form collaboration with others, it is critical for itself to increase its degree of commitment to collaborate and solve problems, and strengthen its organizational capacity by increasing its resources level. Furthermore, to promoate cross sector collaboration, facilitator can play a positive role and increase the number of collaboration formed by pairing up organizations that have relatively high degree of commitment.
